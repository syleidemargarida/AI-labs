```
sentences =[
    "In Mozambique we have many beautiful beaches." ,
    "Mozambique is a PALOP country.",
    "Maputo city is the capital of Mozambique.",

    "Stephen Curry is a good Basketball player.",
    "Barcelona is my favourite team.",
"Artficial intelligence research in India is transforming technology."
"Machine learning and AI are advancing reasearch and innovation."
]
```

```
#importing the libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #TF-Bow
from sklearn.feature_extraction.text import TfidfVectorizer #TFIDF
from sklearn.metrics.pairwise import cosine_similarity
```

```
#Embedding using TF-Bow approach
tf_vectorizer=CountVectorizer()
tf_matrix=tf_vectorizer.fit_transform(sentences)
```

```
tf_matrix
```

```
tf_similarity=cosine_similarity(tf_matrix)
tf_similarity
```

```
#Embedding using TFIDF approach
tfidf_vectorizer=TfidfVectorizer()
tfidf_matrix=tfidf_vectorizer.fit_transform(sentences)
```

```
tfidf_matrix
```

```
tfidf_similarity=cosine_similarity(tfidf_matrix)
tfidf_similarity
```

```
#Findinfout the document similarity
def doc_sim(similarity_matrix,sentences):
  for i in range(len(sentences)):
    sims=similarity_matrix[i].copy()
    sims[i]=-1
    top_index=np.argmax(sims)
    print(sentences[i],"-->", sentences[top_index])
```

```
doc_sim(tfidf_similarity,sentences)
```
