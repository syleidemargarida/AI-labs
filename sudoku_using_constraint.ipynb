```
# Returns a set of valid numbers that can be placed at (row, col)
def get_candidates(mat, row, col):
    # If already filled, no candidates needed
    if mat[row][col] != 0:
        return set()

    # Start with all possible numbers
    nums = set(range(1, 10))

    # Remove numbers already in the same row and column
    for i in range(9):
        nums.discard(mat[row][i])  # row
        nums.discard(mat[i][col])  # column

    # Remove numbers in the 3x3 subgrid
    startRow, startCol = 3 * (row // 3), 3 * (col // 3)
    for i in range(3):
        for j in range(3):
            nums.discard(mat[startRow + i][startCol + j])

    return nums
```

```
# Finds the empty cell with the minimum remaining values (MRV)
def find_mrv_cell(mat):
    min_candidates = 10  # more than max possible (9)
    best_cell = None

    for i in range(9):
        for j in range(9):
            if mat[i][j] == 0:  # only consider empty cells
                candidates = get_candidates(mat, i, j)

                # If no candidates → dead end (fail early)
                if len(candidates) == 0:
                    return (i, j)

                # Pick cell with smallest candidate set
                if len(candidates) < min_candidates:
                    min_candidates = len(candidates)
                    best_cell = (i, j)

                # Best possible case → only 1 option
                if min_candidates == 1:
                    return best_cell

    return best_cell  # None means puzzle is solved
```

```
# Finds the empty cell with the minimum remaining values (MRV)
def find_mrv_cell(mat):
    min_candidates = 10  # more than max possible (9)
    best_cell = None

    for i in range(9):
        for j in range(9):
            if mat[i][j] == 0:  # only consider empty cells
                candidates = get_candidates(mat, i, j)

                # If no candidates → dead end (fail early)
                if len(candidates) == 0:
                    return (i, j)

                # Pick cell with smallest candidate set
                if len(candidates) < min_candidates:
                    min_candidates = len(candidates)
                    best_cell = (i, j)

                # Best possible case → only 1 option
                if min_candidates == 1:
                    return best_cell

    return best_cell  # None means puzzle is solved
def solveSudokuCSP(mat):
    # Select next cell using MRV
    cell = find_mrv_cell(mat)

    # If no empty cell left → solved
    if cell is None:
        return True

    row, col = cell

    # Get valid candidates for this cell
    candidates = get_candidates(mat, row, col)

    # Try each candidate
    for num in candidates:
        mat[row][col] = num  # assign value

        # Recursively solve the rest
        if solveSudokuCSP(mat):
            return True

        # Backtrack if failed
        mat[row][col] = 0

    return False  # trigger backtracking
```

```
def print_sudoku(mat):
    for i, row in enumerate(mat):
        if i % 3 == 0 and i != 0:
            print("-" * 21)  # horizontal divider

        for j, val in enumerate(row):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")

            print(val if val != 0 else ".", end=" ")

        print()
```

```
def solveSudoku(mat):
    if solveSudokuCSP(mat):
        print("Solved Sudoku:\n")
        print_sudoku(mat)
    else:
        print("No solution exists.")
```
